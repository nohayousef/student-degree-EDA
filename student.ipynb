# EDA

```python
gender_grp = df.groupby('Gender')
gender_grp[['AttendanceRate', 'StudyHoursPerWeek','PreviousGrade', 'ExtracurricularActivities', 'FinalGrade']].mean()
```

```python
gender_grp = df.groupby('Gender')
mean_values = gender_grp[['AttendanceRate', 'StudyHoursPerWeek', 'PreviousGrade', 'ExtracurricularActivities', 'FinalGrade']].mean()

# figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))  # the figure size


genders = mean_values.index # get the unique gender values

# create a bar chart for each column
bar_width = 0.1  
x = [i for i in range(len(genders))]  

for i, col in enumerate(mean_values.columns):
    ax.bar([i + j * bar_width for j in range(len(genders))], mean_values[col], width=bar_width, label=col)

# title and labels
ax.set_title('Mean Values by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Mean Value')

# set x ticks
ax.set_xticks([i + (len(genders) - 1) * bar_width / 2 for i in range(len(genders))])
ax.set_xticklabels(genders)

ax.legend()

# show the plot
plt.tight_layout()  
plt.show()
```

```python
unpivot = gender_grp['ParentalSupport'].value_counts(normalize=True).reset_index()
unpivot
```

```python
pivoted_df = unpivot.pivot_table(index='ParentalSupport', values='proportion', columns=unpivot.index)

plt.figure(figsize=(10, 8))
sns.heatmap(pivoted_df, annot=True, cmap='Blues')
plt.xlabel('Parental Support')
plt.ylabel('Gender')
plt.title('Distribution of Parental Support by Gender')
plt.show()
```

```python
# parental support by gender
pivot = unpivot.pivot_table(index='ParentalSupport', columns='Gender', values='proportion', fill_value=0)
pivot
```

```python
# Create a heatmap visualization
plt.figure(figsize=(10, 8))
plt.pcolor(pivot, cmap='Blues')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Parental Support', fontsize=14)
plt.title('Distribution of Parental Support by Gender', fontsize=16)
plt.xticks(np.arange(0.5, len(pivot.columns), 1), pivot.columns, rotation=45, ha='right')
plt.yticks(np.arange(0.5, len(pivot.index), 1), pivot.index, rotation=0, ha='right')
plt.colorbar(label='Proportion', shrink=0.8)
plt.show()
```

```python
import streamlit as st
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
df = px.data.tips()
st.title("Interactive Visualizations")
cat_cols = list(df.select_dtypes(include='O').columns)
num_cols = list(df.select_dtypes(include='number').columns)
num_choice = st.radio(label="Numerical Column", options=num_cols)
cat_choice = st.selectbox(label="Categorical Column", options=cat_cols)
graph_choice = st.selectbox(label="Graph Type", options=['box', 'violin', 'strip'])
choice_to_graph = {'box':px.box, 'violin':px.violin, 'strip':px.strip}

is_submit = st.button("Graph")
if is_submit == True:
    fig = choice_to_graph[graph_choice](df, x=cat_choice, y=num_choice)
    st.plotly_chart(fig)
```
